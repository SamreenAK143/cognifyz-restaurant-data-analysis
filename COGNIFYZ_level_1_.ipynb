#3b)Calculate the percentage of restaurants in each price range category.

```python
price_range_count=df['Price range'].value_counts()
total_rest=len(df['Restaurant ID'])
percentage=round((price_range_count/total_rest)*100,2)
percentage_m=pd.merge(price_range_count,percentage,on="Price range")
print("the percentage of restaurants in each price range category are:",percentage_m)
```

```python
print(price_range_count)
```

```python
plt.pie(percentage, labels=percentage.index, autopct='%1.1f%%',colors=["yellow","green","darkgreen","lightgreen"])
plt.title("percentage of restaurants in price range categories ")
plt.show()
```

#Task4
#4a)Determine the percentage of restaurants that offer online delivery.

```python
print(df['Has Online delivery'].unique())
```

```python
onlin_delivery_count = df['Has Online delivery'].value_counts()
onlin_delivery_count
```

```python
df['Has Online delivery'] = df['Has Online delivery'].str.strip().str.lower() #TO CHANGE ALL TO LOWER CASE IF THERE ARE INCONSISTENT VALUES
print(df['Has Online delivery'])
```

```python
online_delivery = df['Has Online delivery'].value_counts().get('yes',0)
print("total no. of restaurants with online delivery are:", online_delivery)
```

```python
no_online_delivery = df['Has Online delivery'].value_counts().get('no',0)
print("total no. of restaurants with online delivery are:", no_online_delivery)
```

```python
total_rest=len(df['Restaurant ID'])
percent_online_delivery=round((online_delivery/total_rest)*100,2)
print("the percentage of restaurants that offer online delivery is",percent_online_delivery)
```

```python
total_rest=len(df['Restaurant ID'])
percent_no_online_delivery=round((no_online_delivery/total_rest)*100,2)
print("the percentage of restaurants that does not offer online delivery is",percent_no_online_delivery)
```

#4b)Compare the average ratings of restaurants with and without online delivery.

```python
avg_rating_yes=round(df[df['Has Online delivery'] =='yes']['Aggregate rating'].mean(),2)
print("average ratings of restaurants with online delivery is",avg_rating_yes)
```

```python
avg_rating_no=round(df[df['Has Online delivery'] =='no']['Aggregate rating'].mean(),2)
print("average ratings of restaurants without online delivery is",avg_rating_no)
```

```python
labels=['Yes','No']
avg_rating =[avg_rating_yes,avg_rating_no]
colors=['pink','purple']
plt.barh(labels,avg_rating,color=colors )
plt.xlabel('Online Delivery')
plt.ylabel('avg rating')
plt.title('comparing average rating of Online Delivery')
plt.show()
```
